```python
#Loading Libraries
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
#Initializing Raw data into a variable
with open('Watches.txt') as test:
    data = test.read().split("\n")
```

```python
data
```

```python
#Converting Raw Data into a dataframe
df = pd.DataFrame(data, columns = ["COL"])
df.head()
```

```python
#Creating a dictionary to map key value pair
m = df['COL'].str.split(':', expand=True).groupby(0)[1].apply(list).reset_index()
df = pd.DataFrame(dict(zip(m[0], m[1])))
```

```python
df.head(2)
```

```python
df.describe()
```

```python
#Renaming Columns
df.columns = ['NAN','price','productId','title','review','profileName','score','summary','text','date','userId']
df
```

```python
#Converting Unix timestamp into YYYY-MM-DD format
df['purchasedate'] = pd.to_datetime(df['date'],unit= 's')
print(df.groupby( df['purchasedate'].dt.year))
del df['date']
df.head()
```

```python
del df['NAN']
```

```python
df.head(2)#Main Dataset
```

```python
#Data Segregating
selected = df[['productId', 'title', 'profileName','score','purchasedate','summary','userId']]
selected.head(5)
```

```python
print(selected.groupby( selected['purchasedate'].dt.year))
```

```python
#Data splitting
start_date = '2011-01-01'
mask = df[df['purchasedate'] > start_date]
```
